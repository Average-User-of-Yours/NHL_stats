# Notebook Configuration

## Importing Required Libraries

```python
from selenium.webdriver import Chrome
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium import webdriver

import time
import os
import requests
```

## Driver and Selenium Configuration

```python
service = ChromeService(r"..\drivers\chromedriver.exe")
browser = Chrome(service=service)


browser.get("https://www.scrapethissite.com/pages/forms/")
```

# Fetching Website Content

```python
browser.get("https://www.scrapethissite.com/pages/forms/")
result = browser.find_element("class name","table")
source = browser.page_source
```

```python
pages = range(1,25)
output_dir = r"..\data\raw\\"

for page in pages:
    browser.get(f"https://www.scrapethissite.com/pages/forms/?page_num={page}")
    for_check = requests.get(f"https://www.scrapethissite.com/pages/forms/?page_num={page}")
    if for_check.status_code != 404:
        source = browser.page_source
        file_path = os.path.join(output_dir, f"hockey_teams_page_{page}.html")
        with open(file_path, "w", encoding="utf-8") as file:
            file.write(source)
            time.sleep(2)

    else:
        break
```

# Summary

Downloading raw data from our source has reduced the risk of problems stemming from site updates during the extraction process. This method also offers an additional benefit: it allows easy access to the data in its original form, which is crucial if reprocessing is needed.
